## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
#importing numpy and pandas libraries to read the data
import numpy as np
import pandas as pd

#read the csv file and view some sample records
poke_df = pd.read_csv('PokeData_cleaned.csv')
```

```python
poke_df.head()
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
#import matplotlib and seaborn libraries to visualize the data
import matplotlib.pyplot as plt
import seaborn as sns

#Are pokemon more offensive or defensive? 
#using pairplot we'll visualize the data for correlation
sns.pairplot(poke_df, x_vars=['Attack', 'Defense', 'Sp. Atk', 'Sp. Def'], y_vars=['Attack', 'Defense', 'Sp. Atk', 'Sp. Def'], height=4, aspect=1, kind='scatter')
```

```python
#visualizing the data using heatmap
sns.heatmap(poke_df.corr(), cmap='YlGnBu', annot = True)
plt.show()
```

## Create Your Model

Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
#creating x and y
x = poke_df['Attack']
y = poke_df['Defense']
```

```python
#splitting the varaibles as training and testing
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test, = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = 100)

#test
x_train
y_train
```

```python
#import statsmodels.api library from stamodel package
import statsmodels.api as sm

#adding a constant to get an intercept
x_train_sm = sm.add_constant(x_train)

#fitting the regression line using "OLS"
lr = sm.OLS(y_train, x_train_sm).fit()

#print the parameters
lr.params
```

```python
#performing a summary to list out all the different parameters of the regression line fitted.
lr.summary()

#The coefficients and its p-value(significance)
#R-squared value
#F-statistic and its significance
```

```python
#visualize the regression line
plt.scatter(x_train, y_train)
plt.plot(x_train, 37.4733 + 0.405*x_train, 'r')
plt.show()
```

## Error Terms

Finally, plot your error terms!

```python
#Predicting y_value using traing data of x
y_train_pred = lr.predict(x_train_sm)

#creating residuals from the y_train data and predicted y_data
res = (y_train - y_train_pred)

# Plotting the histogram using the residual values
fig = plt.figure()
sns.distplot(res, bins = 15)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('y_train - y_train_pred', fontsize = 15)
plt.show()
```

```python
#looking for any patterns in the residuals
plt.scatter(x_train,res)
plt.show()
```

```python
#adding a constant to x_test
x_test_sm = sm.add_constant(x_test)

# Predicting the y values corresponding to X_test_sm
y_test_pred = lr.predict(x_test_sm)

# Printing first predicted values
y_test_pred
```

```python
#importing r2_square
from sklearn.metrics import r2_score

#checking the R-squred values
r_squared = r2_score(y_test, y_test_pred)
r_squared
```

```python
#visualize the line on the test set
plt.scatter(x_test, y_test)
plt.plot(x_test, y_test_pred, 'r')
plt.show()
```

```python
# splitting the data into train and test
from sklearn.model_selection import train_test_split
x_train_lm, x_test_lm, y_train_lm, y_test_lm = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = 100)

# Shape of the train set without adding column
x_train_lm.shape

# Adding additional column to the train and test data
x_train_lm = x_train_lm.values.reshape(-1,1)
x_test_lm = x_test_lm.values.reshape(-1,1)

print(x_train_lm.shape)
print(x_test_lm.shape)
```

```python
from sklearn.linear_model import LinearRegression

#creating an object of linear regression
lm= LinearRegression()

#fit the model using.fit() method
lm.fit(x_train_lm, y_train_lm)

# Intercept value
print("Intercept :",lm.intercept_)

# Slope value
print('Slope :',lm.coef_)
```

```python
#making predictions of y_value
y_train_pred = lm.predict(x_train_lm)
y_test_pred = lm.predict(x_test_lm)

# Comparing the r2 value of both train and test data
print(r2_score(y_train,y_train_pred))
print(r2_score(y_test,y_test_pred))
```
